Lets try to find and define "cluster"

```python
tumor_file='TCGA_Glioma_HERV52/TCGA-06-0211-01A.c.CT.tri.txt'
normal_file='TCGA_Glioma_HERV52/TCGA-06-0211-10A.c.CT.tri.txt'
```

```python
import pandas
tumor=pandas.read_table(tumor_file,header=None,names=('chr','position','tri')).sort_values(['chr','position'])
normal=pandas.read_table(normal_file,header=None,names=('chr','position','tri')).sort_values(['chr','position'])
```

```python
tumor['chr'].unique()
```

```python
normal['chr'].unique()
```

```python
tumor.loc[0,'d_prev'] = -1
tumor.loc[0]
```

```python
normal.loc[0,'d_prev'] = -1
normal.loc[0]
```

```python
for i in range(1, len(tumor)): # range is not inclusive of the last number
    tumor.loc[i,'d_prev'] = tumor.loc[i,'position'] - tumor.loc[i-1,'position']
```

```python
for i in range(1, len(normal)): # range is not inclusive of the last number
    normal.loc[i,'d_prev'] = normal.loc[i,'position'] - normal.loc[i-1,'position']
```

```python
#pandas.option_context('display.max_rows', None)

tumor.query('d_prev > 0 and d_prev < 50')
```

```python
tumor.loc[len(tumor)-1,'d_next']  = -1
tumor.iloc[-1]
```

```python
normal.loc[len(normal)-1,'d_next']  = -1
normal.iloc[-1]
```

```python
tumor.iloc[-1].d_next 
```

```python
normal.iloc[-1].d_next 
```

```python
for i in range(0, len(tumor)-1): # range is not inclusive of the last number
    tumor.loc[i,'d_next'] = tumor.loc[i+1,'position'] - tumor.loc[i,'position']
```

```python
for i in range(0, len(normal)-1): # range is not inclusive of the last number
    normal.loc[i,'d_next'] = normal.loc[i+1,'position'] - normal.loc[i,'position']
```

```python
t_clusters=tumor.query('d_prev > 0 and d_prev < 40 and d_next > 0 and d_next < 40')
t_clusters
```

```python
n_clusters=normal.query('d_prev > 0 and d_prev < 40 and d_next > 0 and d_next < 40')
n_clusters
```

```python
import numpy
n_array=numpy.array(n_clusters['chr']+':'+n_clusters['position'].astype(str))
t_array=numpy.array(t_clusters['chr']+':'+t_clusters['position'].astype(str))
n_array
```
