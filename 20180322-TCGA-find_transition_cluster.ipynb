```python
f='TCGA_Glioma_HERV52/TCGA-06-0211-10A.c.CT.tri.txt'
```

```python
import pandas
df=pandas.read_table(f,header=None,names=('chr','position','tri'))
df.head()
```

```python
df['chr'].unique()
```

```python
df=df.sort_values(['chr','position'])
```

```python
df.loc[0,'d_prev'] = -1
df.loc[0]
```

```python
for i in range(1, len(df)): # range is not inclusive of the last number
    df.loc[i,'d_prev'] = df.loc[i,'position'] - df.loc[i-1,'position']
```

```python
df.head()
```

```python
#pandas.option_context('display.max_rows', None)

df.query('d_prev > 0 and d_prev < 50')
```

```python
df.loc[len(df)-1,'d_next']  = -1
df.iloc[-1]
```

```python
df.iloc[-1].d_next 
```

```python
for i in range(0, len(df)-1): # range is not inclusive of the last number
    df.loc[i,'d_next'] = df.loc[i+1,'position'] - df.loc[i,'position']
```

```python
clusters = df.query('d_prev > 0 and d_prev < 40 and d_next > 0 and d_next < 40')
clusters
```

So let's define a "cluster" as a position where the one before and the one after are both within 40bp.
