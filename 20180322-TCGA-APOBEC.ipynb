```python
import pandas
```

```python
import glob
#files=glob.glob("*.?.??.txt")
files=glob.glob("TCGA_Glioma_HERV52/*.tri.txt")
print(len(files))
files[:10]
```

```python
import os
#files = [ os.path.basename(f) for f in files ]
```

```python
#df=pandas.DataFrame(columns=['chr','position'])
df=pandas.DataFrame(columns=['chr','position','tri'])
for file in files:
    print(file)
#    r=pandas.read_table(file,header=None,names=('chr','position'))
    r=pandas.read_table(file,header=None,names=('chr','position','tri'))
    r[os.path.basename(file)]=1
    df=pandas.merge(df,r,how='outer')
df.fillna(0,inplace=True)
```

```python
df.head()
```

```python
df.index=df['chr']+':'+df['position'].astype(str)

#df.set_index

#df.drop(columns=['chr','position','tri'],inplace=True)
df.head()
```

```python
df['tn']='normal'
df.loc[df.index.str.match('.{12}-01.*'),'tn'] = 'tumor'
df['tn'].head()
```

```python
df['transition']='CT'
#df.loc[df.index.str.match('.*\.GA\.txt'),'transition'] = 'GA'
df.loc[df.index.str.match('.*\.GA\.tri\.txt'),'transition'] = 'GA'
df['transition'].head()
```

```python
df['caller']='consensus'
#df.loc[df.index.str.match('.*\.m\...\.txt'),'caller'] = 'multiallelic'
df.loc[df.index.str.match('.*\.m\...\.tri\.txt'),'caller'] = 'multiallelic'
df['caller'].head()
```

```python
df['subject'] = df.index.str[0:12]
df['subject'].head()
```

```python
from sklearn.model_selection import train_test_split

for tri in ['ACA','ACC','ACG','ACT','CCA','CCC','CCG','CCT','GCA','GCC','GCG','GCT','TCA','TCC','TCG','TCT',
            'AGA','AGC','AGG','AGT','CGA','CGC','CGC','CGT','GGA','GGC','GGG','GGT','TGA','TGC','TGG','TGT']:
    sub=df.transpose().query('tri == "'+tri+'"').transpose()
    # that unfortunately chops of my added columns 'tn','transition','caller','subject'

    sub['tn']='normal'
    sub.loc[sub.index.str.match('.{12}-01.*'),'tn'] = 'tumor'

    X = sub.drop(['chr','position','tri']).drop(columns=['tn'])
    y = sub.drop(['chr','position','tri']).tn
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.4)
    logreg.fit(X_train,y_train)
    y_pred = logreg.predict(X_test)
    print( tri + " : " + str(metrics.accuracy_score(y_test,y_pred)) ) 
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=9)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.4)
logreg.fit(X_train,y_train)
y_pred = logreg.predict(X_test)
print(metrics.accuracy_score(y_test,y_pred))
```

```python
k_range = list(range(1, 50))
scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    scores.append(metrics.accuracy_score(y_test, y_pred))
    
# import Matplotlib (scientific plotting library)
import matplotlib.pyplot as plt

# allow plots to appear within the notebook
%matplotlib inline

# plot the relationship between K and testing accuracy
plt.plot(k_range, scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Testing Accuracy')
0
```
