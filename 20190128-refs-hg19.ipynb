
What portions of each virus have some heterogeneity to the hg19 human reference?

I broke up hg19 into 100bp chunks and aligned to each viral reference.


```
faSplit size hg19.fa.gz 100 OUTDIR

cat OUTDIR/*fa | gzip > hg19.100.fa.gz

bowtie2 --all --no-unal --threads 35 -f --xeq -x virii/${virus} --very-sensitive-local -U hg19.100.fa.gz | samtools view -F 4 -o ${virus}.unsorted.bam -

samtools sort -o ${virus}.bam ${virus}.unsorted.bam

samtools depth ${virus}.bam | awk -f depths_to_reverse_regions.awk > ${virus}.nonhg19.txt

samtools depth ${virus}.bam > ${virus}.hg19.depth.txt

```




````python
import glob
import os.path
import pandas
import matplotlib.pyplot as plt
````

````python
# For some reason, if this is in the same cell as the import command, it doesn't work.
plt.rcParams["figure.figsize"] = [18.0,8.0]
````

````python
virii=pandas.read_csv('/Users/jakewendt/20181214-1000genomes-unmapped-virii/virii_details.csv', sep="\t", header=None, 
                      names=["accession","description"],
                      index_col=["accession"])
virii.head()
````

````python
for filepath in sorted(glob.iglob('/Users/jakewendt/20190128-refs-hg19/*hg19.depth.txt')):
    #print(filepath)
    virus=os.path.basename(filepath)
    virus = virus[:virus.index('.',virus.index('.')+1)]
    print(virus + " - " + virii.loc[virus].description)
    if os.path.isfile(filepath): 
        d = pandas.read_csv(filepath,
            sep="\t",
            header=None,
            usecols=[1,2],
            names=["position","depth"],
            index_col=["position"] )
        d.plot( title="Depth chart of 100bp chunks of hg19 mapped to "+virus, y='depth',ylim=[-1,100])
        plt.show()
````
